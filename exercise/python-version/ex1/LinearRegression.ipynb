# 线性回归

## 简单线性回归

```python
from pandas import DataFrame, Series
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. 数据可视化——一图胜千言

```python
data1 = pd.read_table('ex1data1.txt', sep=',', header=None)
data1.columns = ['area', 'price']
m, n = data1.shape
```

```python
%matplotlib inline
plt.plot( data1.area, data1.price, 'rx')
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
plt.title('Profit vs. Population')
plt.show()
```

### 2. 梯度下降法求参数

```python
X = data1.area
y = data1.price
X = X.reshape(len(X), 1)
X = np.hstack((np.ones((m, 1)), X))
theta = np.zeros(n)
iterations = 1500
alpha = 0.01
```

```python
def compute_cost(X, y, theta):
    m = float(len(y))
    res = np.dot(X, theta) - y
    return np.dot(res.T, res) / (2*m)

def gradient_descent(X, y, theta, alpha, num_iters):
    m = float(len(y))
    J_hist = []
    
    for i in range(num_iters):
        theta = theta - alpha * np.dot(X.T, np.dot(X, theta) - y) / m
        J_hist.append(compute_cost(X, y, theta))
    
    return theta, J_hist
```

```python
theta, J_hist = gradient_descent(X, y, theta, alpha, iterations)
```

```python
plt.plot(data1.area, data1.price, 'rx')
plt.plot(data1.area, np.dot(X, theta), 'b-')
plt.show()
```

```python
plt.plot(J_hist)
plt.ylim(min(J_hist)-0.1, max(J_hist[1:])+0.1)
plt.title('$J(\\theta)$ by iterations')
plt.ylabel('$J(\\theta)$')
plt.xlabel('# of iterations')
plt.show()
```

## 多元线性回归

```python
data2 = pd.read_table('ex1data2.txt', sep=',', header=None)
data2.columns = ['area', 'num_rooms','price']
m, n = data2.shape
```

### Gradient Descent

```python
X = data2[['area', 'num_rooms']]
y = data2['price']
```

```python
# feature normalize
def feature_normalize(X):
    return (X - X.mean(axis=0)) / X.std(axis=0)

X = feature_normalize(X)
X = np.hstack((np.ones((m, 1)), X))
```

```python
theta = np.zeros(n)
```

```python
# gradient descnet
alpha = 1
num_iters = 400
theta, J_hist = gradient_descent(X, y, theta, alpha, num_iters)
print theta
```

### $\alpha$ 选择

```python
alphas = [0.01, 0.03,0.1,0.3, 1, 3, 10]

for alpha in alphas:
    theta = np.zeros(n)
    theta, J_hist = gradient_descent(X, y, theta, alpha, num_iters)
    plt.plot(J_hist, 'r-')
    plt.title('$\\alpha$=%.2f' % alpha)
    plt.show()
```

### Normal Equation

```python
X = data2[['area', 'num_rooms']]
y = data2['price']
X = np.hstack((np.ones((m, 1)), X))
```

```python
theta = np.dot(np.dot(np.linalg.pinv(np.dot(X.T, X)), X.T), y)
```

```python
def ols(X, y):
    return np.dot(np.dot(np.linalg.pinv(np.dot(X.T, X)), X.T), y)

theta_ne = ols(X, y)
```
